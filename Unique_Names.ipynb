```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:\Studies\ML\challenge.csv')
df.sort_values(['fn','ln'],inplace=True)
df.reset_index(drop = True, inplace = True)
```

```python
x = df.fn.drop_duplicates()
x.reset_index(drop = True, inplace = True)
x = x.to_frame('ufn')
y = x.copy(deep = True)
```

```python
i = 0
while(i<len(x.index)):
    t = x.ufn.str.contains(x.loc[i,'ufn'])
    p = t.to_frame('ufn')
    j = i+1
    while(j<len(x.index)):
        if p.loc[j,'ufn'] == True:
            y.drop([j],inplace = True)
            j = j+1
            i = i+1
        else:
            j =j+1
    i = i+1
y.reset_index(drop = True,inplace = True)
y[:]
```

```python
d = df[:]['dob'] +' '+ df[:]['gn']
d = d.to_frame('pt')
d= d.pt.str.split(pat = ' ',expand = True)
d['ln'] = df[:]['ln']
d['fn'] = df[:]['fn']
d['fln'] = df[:]['fn'] + ' '+ df[:]['ln']
d_cols = ['dob','gn','ln','fn','fln']
d.columns = d_cols
d.drop_duplicates(subset = ['fln','dob'],inplace = True)
d.reset_index(drop = True, inplace = True)
d.head()
```

```python
a = pd.DataFrame(columns = ['fln','ln','dob'])  #final dataframe 
```

```python
b = 0
a = pd.DataFrame(columns = ['fln','ln','dob'])  #final dataframe
while(b<len(y.index)):
    ab = d.fn.str.contains(y.loc[b,'ufn'])
    a_b = ab.to_frame('noc')
    k = 0
    c = pd.DataFrame(columns = ['fln','ln','dob'])
    e = pd.DataFrame(columns = ['fln','ln','dob'])
    while(k<len(a_b.index)):
        if a_b.loc[k,'noc'] == True:
            c = c.append(d.loc[k,['fln','ln','dob']])
            k = k+1
        else:
            k = k+1
    c.reset_index(drop = True,inplace = True)
    e = c.copy(deep = True)
    if len(e.index) == 1:
        a = a.append(e,ignore_index = True)
        b = b+1
    else:
        f = 0
        while(f<len(c.index)):
            mn = c.ln.str.contains(c.loc[f,'ln'])
            m_n = mn.to_frame('noc')
            q = f+1
            while(q<len(c.index)):
                if (m_n.loc[q,'noc'] == True) & (c.loc[q-1,'dob'] == c.loc[q,'dob']):
                    e.drop([q],inplace = True)
                    q = q +1
                    f = f+1
                else:
                    q = q+1
            f = f+1
        e.reset_index(drop = True,inplace = True)
        a = a.append(e,ignore_index = True)
        b =b+1        
a.reset_index(drop = True,inplace = True)
```
